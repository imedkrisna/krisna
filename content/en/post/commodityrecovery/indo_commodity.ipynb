```python
import quandl
quandl.ApiConfig.api_key = "vDm7zkSvztuywGuZuuvY"
mydata = quandl.get("UCOM/NICK_IDN", paginate=True)
mydata = mydata.reset_index()
mydata['Nickel price (USD/Kg)']=mydata['Nickel mattes - Export - Trade (USD)']/mydata['Nickel mattes - Export - Weight (Kg)']
mydata
```
