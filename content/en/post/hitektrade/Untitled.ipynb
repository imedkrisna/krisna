```python
import pandas as pd
pd.set_option('max_rows', None)
import pandas as pd
import datetime
import seaborn as sns; sns.set()
import matplotlib.pyplot as plt
a=pd.read_csv("pma.csv")
a=a.fillna(0) # replace NaN with zero
b=list(a.columns.values)
del b[0]
a=a.melt(id_vars='PMA',value_vars=b) # melt so time columns become 1 columns only
a=a.rename(columns={"variable" : "year", "value" : "kUSD"})
a['year']=pd.to_datetime(a['year']) # change variable to time variable
z=a.PMA.unique() # bikin list buat jenis PMA
a=a.set_index('PMA')
#b=a[a['PMA']==z[27]]
#b=b.set_index('PMA')
b=a.loc[[z[27],z[54]]]
elektronik=a.loc[z[27]]
elektronik=elektronik.rename(columns={'kUSD' : 'electronics (right)'})
it=a.loc[z[54]]
it=it.rename(columns={'kUSD' : 'software (left)'})
itu=a.loc[z[55]]
itu=itu.rename(columns={'kUSD' : 'web based (right)'})
b=b.reset_index()
plt.figure(figsize=(7,5))
#fuk=sns.lineplot(x="year", y="kUSD",hue="PMA",data=b)
#fuk.axvline(datetime.datetime(2016,1,1), color='red')
#fuk.text(datetime.datetime(2013,1,1),500000, "LCR",color='brown',size='x-large')
ax = it.plot(x="year", y="software (left)", legend=False)
ax2 = ax.twinx()
elektronik.plot(x="year", y="electronics (right)", ax=ax2, legend=False, color="r")
itu.plot(x="year", y="web based (right)", ax=ax2, legend=False, color="orange")
ax.axvline(datetime.datetime(2016,1,1), color='black')
ax.text(datetime.datetime(2013,1,1),30000, "LCR",color='brown',size='x-large')
ax.figure.legend(loc=9)
plt.show()
```
